```python
housing = pd.read_csv("data.csv")
```

```python
housing['num_years_since_built'] = 2023 - housing['yr_built']
housing['num_yrs_since_ren'] = 0
for i in range(len(housing)):
    if housing["yr_renovated"].iloc[i] != 0:
        housing.at[i, 'num_yrs_since_ren'] = 2023 - housing["yr_renovated"].iloc[i]
housing = housing.drop(["yr_built", "yr_renovated"], axis=1)
housing.head(5)
```

```python
# pd.set_option('display.max_rows',None)
housing.isnull().sum().sum()
```

```python
missing_values = [feature for feature in housing.columns if housing[feature].isnull().sum() == 0]
for f in missing_values:
    print("{} does not have missing values.".format(f))
```

```python
import seaborn as sns
sns.heatmap(housing.isnull(),yticklabels=False,cbar=False)
```

```python
import matplotlib.pyplot as plt
for feature in missing_values:
    d=housing.copy()
    d[feature]=np.where(d[feature].isnull(),1,0)
    d.groupby(feature)["price"].median().plot.bar()
    plt.show()
```

```python
housing["view"].value_counts()
```

DISCRETE FEATURES


```python
numerical_features = [feature for feature in housing.columns if housing[feature].dtypes != 'O'] 
print(numerical_features)


discrete_feature=[feature for feature in numerical_features if len(housing[feature].unique())<25 ]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
for feature in discrete_feature:
    data=housing.copy()
    data.groupby(feature)['price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('price')
    plt.title(feature)
    plt.show()
```

CONTINUOUS FEATURES

```python
housing=housing.drop(["date","statezip","street","city","country"],axis=1)
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

OUTLIERS

```python
import seaborn as sns
for feature in continuous_feature:
    data=housing.copy()
    if 0 in data[feature].unique():
        pass
    else:
#     sns.boxplot(housing[feature])
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
# pd.set_option("display.max_rows",None)
correlation_matrix = housing.corr().abs()
correlation_with_price = correlation_matrix['price'].sort_values(ascending=False)
print(correlation_with_price)
```

```python
len(correlation_with_price)
```

```python
correlations = housing.corr().abs()['price']
columns_to_drop = correlations[correlations < 0.05].index.tolist()
# columns_to_drop = correlations[correlations < 0.35].index.tolist()
housing = housing.drop(columns=columns_to_drop, axis=1)
```

```python
import seaborn as sns
corr_matrix = housing.corr().round(1)
sns.heatmap(data=corr_matrix, annot=True)
```

```python
from pandas.plotting import scatter_matrix
attributes = ["price", "sqft_living", "sqft_above","bathrooms"]
scatter_matrix(housing[attributes], figsize = (12,8))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))

features = ["sqft_living", "sqft_above"]
target = housing["price"]

for i, col in enumerate(features):
    plt.subplot(1,len(features), i+1)
    x = housing[col]
    y = target
    plt.scatter(x,y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('price')
```
